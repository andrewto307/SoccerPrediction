```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 200)
```

# <h1 style='font-size:30px;'>Data Collecting</h1>


```python
training_dictionary = { "df_0809": "../data/08-09.csv",
                        "df_0910": "../data/09-10.csv",
                        "df_1011": "../data/10-11.csv",
                        "df_1112": "../data/11-12.csv",
                        "df_1213": "../data/12-13.csv",
                        "df_1314": "../data/13-14.csv", 
                        "df_1415": "../data/14-15.csv",
                        "df_1516": "../data/15-16.csv",
                        "df_1617": "../data/16-17.csv",
                        "df_1718": "../data/17-18.csv",
                        "df_1819": "../data/18-19.csv"
                        }

training_dataframes = {}
for key, value in training_dictionary.items():
    training_dataframes[key] = pd.read_csv(value)

dataframes_to_concat = [training_dataframes[key] for key in training_dataframes.keys()]
training_dataset = pd.concat(dataframes_to_concat, axis=0, ignore_index=True)

testing_dataset = pd.read_csv("../data/19-20.csv")
```

# <h1 style='font-size:30px;'>NaN Value Handling</h1>

```python
columns_have_na_values = training_dataset.columns[training_dataset.isna().any()]
```

```python
columns_have_na_values
```

```python
import yaml

with open('../data/description.yaml', 'r') as file:
    column_definitions = yaml.safe_load(file)

for col in columns_have_na_values:
    print(col + ": " + column_definitions.get(col, "No definition"))
```

```python
# Replace NaN values in companies' betting odd by the avg betting odd

for col in ["GBH", "IWH", "LBH", "SBH", "PSH", "SJH", "VCH", "BSH", "PSCH"]:
    training_dataset[col] = training_dataset[col].fillna(training_dataset["BbAvH"])

for col in ["GBD", "IWD", "LBD", "SBD", "PSD", "SJD", "VCD", "BSD", "PSCD"]:
    training_dataset[col] = training_dataset[col].fillna(training_dataset["BbAvD"])

for col in ["GBA", "IWA", "LBA", "SBA", "PSA", "SJA", "VCA", "BSA", "PSCA"]:
    training_dataset[col] = training_dataset[col].fillna(training_dataset["BbAvA"])
```

# <h1 style='font-size:30px;'>Column Alignment</h1>

```python
# Remove the "Bb" (Betbrain) from training dataset's columns that start with it
# Rename columns for alignment

training_dataset.columns =  training_dataset.columns.str.replace(r'^Bb', '', regex=True)
training_dataset.columns = training_dataset.columns.str.replace('Mx', 'Max', regex=False).str.replace('Av', 'Avg', regex=False)

testing_dataset.columns =  testing_dataset.columns.str.replace(r'^Bb', '', regex=True)
testing_dataset.columns = testing_dataset.columns.str.replace('Mx', 'Max', regex=False).str.replace('Av', 'Avg', regex=False)
```

```python
common_columns = training_dataset.columns.intersection(testing_dataset.columns)
training_dataset = training_dataset[common_columns]
testing_dataset = testing_dataset[common_columns]
```
